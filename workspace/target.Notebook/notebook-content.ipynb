```python
import json

from mxdataspark import MXData
```

```python
from importlib.metadata import version
print(f'MXData Spark package version: {version("mxdataspark")}')
```

```python
# DBTITLE 1,Widget Parameters
dbutils.widgets.text("target_type", "")
dbutils.widgets.text("source_catalog", "")
dbutils.widgets.text("source_schema", "")
dbutils.widgets.text("source_table", "")
dbutils.widgets.text("where_clause", "")
dbutils.widgets.text("target_name", "")
dbutils.widgets.text("keyvault_secret_name", "")
dbutils.widgets.text("options", "{}")
```

```python
target_type = dbutils.widgets.get("target_type")
source_catalog = dbutils.widgets.get("source_catalog")
source_schema = dbutils.widgets.get("source_schema")
source_table = dbutils.widgets.get("source_table")
where_clause = dbutils.widgets.get("where_clause")
target_name = dbutils.widgets.get("target_name")
keyvault_secret_name = dbutils.widgets.get("keyvault_secret_name")
options = json.loads(dbutils.widgets.get("options"))

options["source_catalog"] = source_catalog
options["where_clause"] = where_clause
source_table_name = f"{source_schema}.{source_table}"

spark.catalog.setCurrentCatalog(source_catalog)
```

```python
# DBTITLE 1,Main
dbutils.notebook.exit(
    MXData.target(target_type)
    .keyvault_secret(keyvault_secret_name)
    .source_table(source_table_name)
    .target_name(target_name)
    .options(**options)
    .publish()
    .metrics()
)
```
