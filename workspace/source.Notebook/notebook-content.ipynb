```python
import json

from mxdataspark import MXData
```

```python
# DBTITLE 1,Widget Parameters
dbutils.widgets.text("source_type", "")
dbutils.widgets.text("keyvault_secret_name", "")
dbutils.widgets.text("extract_mode", "")
dbutils.widgets.text("output_path", "")
dbutils.widgets.text("object_name", "")
dbutils.widgets.text("object_schema", "")
dbutils.widgets.text("columns", "")
dbutils.widgets.text("column_types", "")
dbutils.widgets.text("options", "")
```

```python
source_type = dbutils.widgets.get("source_type")
keyvault_secret_name = dbutils.widgets.get("keyvault_secret_name")
extract_mode = dbutils.widgets.get("extract_mode")
output_path = dbutils.widgets.get("output_path")
object_name = dbutils.widgets.get("object_name")
object_schema = dbutils.widgets.get("object_schema")
col_names = dbutils.widgets.get("columns")
col_types = dbutils.widgets.get("column_types")
options = json.loads(dbutils.widgets.get("options"))
```

```python
# DBTITLE 1,Main
dbutils.notebook.exit(
    MXData.source(source_type)
    .keyvault_secret(keyvault_secret_name)
    .set_extract_mode(extract_mode)
    .output_path(output_path)
    .object_name(object_name, object_schema)
    .source_columns(col_names, col_types)
    .options(**options)
    .acquire()
    .metrics()
)
```
