```python
import json

from mxdataspark import MXData
```

```python
from importlib.metadata import version
print(f'MXData Spark package version: {version("mxdataspark")}')
```

```python
# DBTITLE 1,Widget Parameters
dbutils.widgets.text("source_file", "")
dbutils.widgets.dropdown("source_format", "parquet", ["parquet", "csv"])
dbutils.widgets.text("source_options", "{}")
dbutils.widgets.text("source_schema", "")
dbutils.widgets.text("delta_schema_name", "")
dbutils.widgets.text("delta_table_name", "")
dbutils.widgets.dropdown("delta_merge_type", "full", ["full", "incremental"])
dbutils.widgets.text("delta_candidate_keys", "")
dbutils.widgets.text("pre_processor", "")
dbutils.widgets.text("catalog_name", "")
dbutils.widgets.text("watermark_type", "")
```

```python
delta_merge_type = dbutils.widgets.get("delta_merge_type")
source_file = "/mnt" + dbutils.widgets.get("source_file")
source_format = dbutils.widgets.get("source_format")
source_options = json.loads(dbutils.widgets.get("source_options") or "{}")
delta_schema_name = dbutils.widgets.get("delta_schema_name")
delta_table_name = dbutils.widgets.get("delta_table_name")
target_table = f"{delta_schema_name}.{delta_table_name}"

configuration = {
    "source_schema": dbutils.widgets.get("source_schema"),
    "delta_candidate_keys": dbutils.widgets.get("delta_candidate_keys"),
    "pre_processor": dbutils.widgets.get("pre_processor"),
    "watermark_type": dbutils.widgets.get("watermark_type"),
}

catalog_name = dbutils.widgets.get("catalog_name").lower()
spark.catalog.setCurrentCatalog(catalog_name)
```

```python
# DBTITLE 1,Main
dbutils.notebook.exit(
    MXData.engine(delta_merge_type)
    .source_data(source_file, source_format, source_options)
    .target_table(target_table)
    .ingest(configuration)
    .metrics()
)
```
